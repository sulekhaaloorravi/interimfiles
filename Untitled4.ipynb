Implementation

```python
#Same of additive and multiplicative
def initiate_trend_factor(sparkDF,L):
    observed = [p.ObservedVals for p in sparkDF.select("ObservedVals").collect()]
    b_numerator = 0.0
    for i in range(L):
        b_numerator += float(observed[L+i] - observed[i])/L
    trendfactor = b_numerator/L
    return trendfactor     
```

```python
def initiate_seasonal_indices_multiplicative(sparkDF,L):
    observed = [p.ObservedVals for p in sparkDF.select("ObservedVals").collect()]  
    period = int(len(observed)/L)
    #Average
    Avgs = []
    for i in range(period):
        Avgs.append(sum(observed[L*i:L*i+L])/float(L))
    #Avgs
    #Subset Values according to season length
    Subset = []
    for i in range(period):
        Subset.append(observed[L*i:L*i+L])
    #Subset
    #Indices
    Indices = []
    for j in range(L):
        for i in range(period):
            Indices.append(float(Subset[i][j]/Avgs[i])) #Divide in multiplicative
    #Indices 
    #SeasonalIndices
    SeasonalIndices = []
    for i in range(L):
        SeasonalIndices.append(float(sum(Indices[period*i:period*i+period])/period))
    return SeasonalIndices
```

```python
def initiate_seasonal_indices_additive(sparkDF,L):
    observed = [p.ObservedVals for p in sparkDF.select("ObservedVals").collect()]  
    period = int(len(observed)/L)
    #Average
    Avgs = []
    for i in range(period):
        Avgs.append(sum(observed[L*i:L*i+L])/float(L))
    #Avgs
    #Subset Values according to season length
    Subset = []
    for i in range(period):
        Subset.append(observed[L*i:L*i+L])
    #Subset
    #Indices
    Indices = []
    for j in range(L):
        for i in range(period):
            Indices.append(float(Subset[i][j]-Avgs[i])) #Subtract in Additive
    #Indices 
    #SeasonalIndices
    SeasonalIndices = []
    for i in range(L):
        SeasonalIndices.append(float(sum(Indices[period*i:period*i+period])/period))
    return SeasonalIndices
```

```python
def initial_seasonal_components(series, slen):
    seasonals = {}
    season_averages = []
    n_seasons = int(len(series)/slen)
    # compute season averages
    for j in range(n_seasons):
        season_averages.append(sum(series[slen*j:slen*j+slen])/float(slen))
    # compute initial values
    for i in range(slen):
        sum_of_vals_over_avg = 0.0
        for j in range(n_seasons):
            sum_of_vals_over_avg += series[slen*j+i]/season_averages[j]
        seasonals[i] = sum_of_vals_over_avg/n_seasons
    return seasonals
```

```python
#Calculate Level, Trend and Seasonality
alpha = 0.2
beta = 0.2
gamma = 0.2
#Observed value at time t
Level = [Avgs[0]]
Trend = [trendfactor]
Seasonality = SeasonalIndices
        
for i in range(len(observed)):
        Level.append(alpha*(observed[i]/Seasonality[i])+(1-alpha)*(Level[i]+Trend[i]))
        Trend.append(beta*(Level[i] - Level[i-1]) + (1-beta)*Trend[i-1])
        Seasonality.append(gamma*(observed[i]/Level[i]) + (1-gamma)*Seasonality[i])
n_predictions = 3
Predictions = []
m = 1
for j in range(len(observed)+n_predictions):
    if j<=len(observed):
        Predictions.append((Level[j] + m*Trend[j]) * Seasonality[j-m])   
    else:
        for m in range(1,n_predictions):
            Predictions.append((Level[len(observed)] + m*Trend[len(observed)]) * Seasonality[j-m])
```
